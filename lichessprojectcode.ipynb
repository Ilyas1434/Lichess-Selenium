```python
import pandas as pd #dataframe module
import numpy as np #numbers module
from selenium import webdriver #imports selenium
from selenium.webdriver.common.keys import Keys #For sending keys
import time #time module for waits
from selenium.webdriver.common.by import By #For explicit wait, and for finding elements
from selenium.webdriver.support.ui import WebDriverWait #For explicit wait
from selenium.webdriver.support import expected_conditions as EC #For explicit wait
from selenium.webdriver import ActionChains
df = pd.read_csv("lichess_db_puzzle.csv")
df.columns = ["PuzzleId","FEN","Moves","Rating","RatingDeviation","Popularity","NbPlays","Themes","GameUrl"]
df[:10]
```

```python

PATH = "C:\Program Files (x86)\chromedriver.exe"
driver = webdriver.Chrome(PATH)
driver.get('https://lichess.org/training')
action_chains = ActionChains(driver)
driver.set_window_size(1050, 652)
for i in range(10):
    puzzle_id_raw = driver.find_element(By.PARTIAL_LINK_TEXT, "#")
    puzzle_id_raw = puzzle_id_raw.get_attribute('innerHTML')
    puzzle_id = puzzle_id_raw[1:]
    selection = df.loc[df['PuzzleId'] == puzzle_id]
    selection = selection["Moves"].tolist()
    selection = selection[0]
    selection = selection[5:].split()
    selection = selection[::2]
    print(selection)
    colorBeingPlayed = driver.find_element(By.TAG_NAME, "em") 
    colorBeingPlayed = colorBeingPlayed.get_attribute('innerHTML')
    colorBeingPlayed = colorBeingPlayed[23:]
    colorBeingPlayed = colorBeingPlayed[:5] 
    print(colorBeingPlayed)
    time.sleep(1)
    Board = driver.find_element(By.TAG_NAME, "cg-board")
    
    size = Board.size
    length = size['width']
    offset = length/8
    print(offset)
    if colorBeingPlayed == "white":
        a = offset
        b = offset*2
        c = offset*3
        d = offset*4
        e = offset*5
        f = offset*6
        g = offset*7
        h = offset*8
        sq1 = offset*7
        sq2 = offset*6
        sq3 = offset*5
        sq4 = offset*4
        sq5 = offset*3
        sq6 = offset*2
        sq7 = offset*1
        sq8 = offset*0
        alphabet = ["a", "b", "c", "d", "e", "f", "g", "h"]
        alphabetVariables = [a,b,c,d,e,f,g,h]
        numbers = ["1", "2", "3", "4", "5", "6", "7", "8"]
        numbersVariables = [sq1,sq2,sq3,sq4,sq5,sq6,sq7,sq8]
        for element in selection:
            firstClick = element[:2]
            firstClickLetter = firstClick[:1]
            i = 0
            for letter in alphabet:
                if firstClickLetter == letter:
                    firstClickLetter = alphabetVariables[i]
                i = i + 1
            firstClickNumber = firstClick[1:]
            j = 0
            for number in numbers:
                if firstClickNumber == number:
                    firstClickNumber = numbersVariables[j]
                j = j + 1
            action_chains.move_to_element_with_offset(Board, firstClickLetter, firstClickNumber).click().perform()

            secondClick = element[2:]
            secondClickLetter = secondClick[:1]
            k = 0
            for letter in alphabet:
                if secondClickLetter == letter:
                    secondClickLetter = alphabetVariables[k]
                k = k + 1
            secondClickNumber = secondClick[1:]
            l = 0
            for number in numbers:
                if secondClickNumber == number:
                    secondClickNumber = numbersVariables[l]
                l = l + 1
            action_chains.move_to_element_with_offset(Board, secondClickLetter, secondClickNumber).click().perform()
    if colorBeingPlayed == "black":
        a = offset*8
        b = offset*7
        c = offset*6
        d = offset*5
        e = offset*4
        f = offset*3
        g = offset*2
        h = offset*1
        sq1 = offset*0
        sq2 = offset*1
        sq3 = offset*2
        sq4 = offset*3
        sq5 = offset*4
        sq6 = offset*5
        sq7 = offset*6
        sq8 = offset*7
        alphabet = ["a","b","c","d","e","f","g","h"]
        alphabetVariables = [a,b,c,d,e,f,g,h]
        numbers = ["1","2","3","4","5","6","7","8"]
        numbersVariables = [sq1,sq2,sq3,sq4,sq5,sq6,sq7,sq8]
        for element in selection:
            firstClick = element[:2]
            firstClickLetter = firstClick[:1]
            i = 0
            for letter in alphabet:
                if firstClickLetter == letter:
                    firstClickLetter = alphabetVariables[i]
                i = i + 1
            firstClickNumber = firstClick[1:]
            j = 0
            for number in numbers:
                if firstClickNumber == number:
                    firstClickNumber = numbersVariables[j]
                j = j + 1
            print(firstClickLetter, firstClickNumber)
            action_chains.move_to_element_with_offset(Board, firstClickLetter, firstClickNumber).click().perform()
            secondClick = element[2:]
            secondClickLetter = secondClick[:1]
            k = 0
            for letter in alphabet:
                if secondClickLetter == letter:
                    secondClickLetter = alphabetVariables[k]
                k = k + 1
            secondClickNumber = secondClick[1:]
            l = 0
            for number in numbers:
                if secondClickNumber == number:
                    secondClickNumber = numbersVariables[l]
                l = l + 1
            print(secondClickLetter, secondClickNumber)
            action_chains.move_to_element_with_offset(Board, secondClickLetter, secondClickNumber).click().perform()
    time.sleep(1)
    next_puzzle = driver.find_element(By.CLASS_NAME, "continue")
    next_puzzle.click()
```
